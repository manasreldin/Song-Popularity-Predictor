```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("SimpleFeatures.csv")
```

```python
data.head()
```

```python
billboard = pd.read_csv("HotSongsBillBoard.csv")
```

```python
billboard.head()
```

```python
bb_tracks = list(billboard.track)

def find_target(row):
    lst = bb_tracks
    if row['title'] in lst:
        return 1
    else:
        return 0
```

```python
data['bbhot'] = data.apply(find_target, axis=1)
```

```python
data['bbhot'].value_counts()
```

```python
def text_clean(x):
    try:
        x = x.decode('utf-8')
    except:
        None
    return x

data = data.applymap(text_clean)
```

```python
data.to_csv("SimpleDataSetWithNewTarget.csv",encoding="utf-8", index = False)
```
